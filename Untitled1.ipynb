```
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

from sklearn.preprocessing import LabelEncoder
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from yellowbrick.cluster import KElbowVisualizer
import os
```

```
names = ['q1', 'q2', 'q3', 'q4', 'q5', 'q6', 'q7', 'q8', 'q9', 'q10', 'q11',
       'q12', 'q13', 'q14', 'q15', 'q16', 'q17', 'q18', 'q19', 'q20', 'q21',
       'q22', 'q23', 'q24', 'q25', 'q26', 'q27', 'q28', 'q29', 'q30', 'q31',
       'q32', 'q33', 'q39', 'q45', 'q46', 'q47', 'q48', 'q49', 'q50', 'q51',
       'q52', 'q53', 'q54']
dataset=pd.read_csv("./dataset.csv" )
```

```
finaldata= pd.DataFrame()
for col in dataset.columns[0:54] :
  if dataset[col].isnull().sum() < 500 :
    finaldata[col] = dataset[col]
  


finaldata = finaldata.dropna(axis=0 , how='any') 

finaldata

```

```

pca = PCA(n_components=2)

reduced_data=pca.fit_transform(finaldata)
plt.scatter(reduced_data[:,0],reduced_data[:,1])
plt.show()
```

```
from sklearn.manifold import TSNE
tsne = TSNE(random_state=17)

data_tsne = tsne.fit_transform(finaldata)

plt.figure(figsize=(12,10))
plt.scatter(data_tsne[:, 0], data_tsne[:, 1], 
            edgecolor='none', alpha=0.7, s=40,
            cmap=plt.cm.get_cmap('nipy_spectral', 10))
plt.colorbar()
plt.title('MNIST. t-SNE projection');
```
