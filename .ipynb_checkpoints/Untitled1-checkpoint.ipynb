```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

from sklearn.preprocessing import LabelEncoder
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA

from sklearn.metrics import silhouette_score
import os
```

```python
pd.set_option('display.max_rows', 120,'display.max_columns', None)
```

```python
names = ['q1', 'q2', 'q3', 'q4', 'q5', 'q6', 'q7', 'q8', 'q9', 'q10', 'q11',
       'q12', 'q13', 'q14', 'q15', 'q16', 'q17', 'q18', 'q19', 'q20', 'q21',
       'q22', 'q23', 'q24', 'q25', 'q26', 'q27', 'q28', 'q29', 'q30', 'q31',
       'q32', 'q33', 'q39', 'q45', 'q46', 'q47', 'q48', 'q49', 'q50', 'q51',
       'q52', 'q53', 'q54']
dataset=pd.read_csv("./dataset.csv" )
```

```python
finaldata= pd.DataFrame()
for col in dataset.columns[0:54] :
  if dataset[col].isnull().sum() < 500 :
    finaldata[col] = dataset[col]
  


finaldata = finaldata.dropna(axis=0 , how='any') 

finaldata

```

```python

pca = PCA(n_components=2)

reduced_data=pca.fit_transform(finaldata)
plt.scatter(reduced_data[:,0],reduced_data[:,1])
plt.show()
```

```python
from sklearn.manifold import TSNE
tsne = TSNE(random_state=17)

result_tsne = tsne.fit_transform(finaldata)
tsne_df_scale = pd.DataFrame(result_tsne, columns=['tsne1', 'tsne2'])
plt.figure(figsize=(12,10))
plt.scatter(result_tsne[:, 0], result_tsne[:, 1], 
            edgecolor='none', alpha=0.7, s=40,
            cmap=plt.cm.get_cmap('nipy_spectral', 10))

plt.xlabel('tsne1')
plt.ylabel('tsne2')

plt.title('MNIST. t-SNE projection');
```

```python
sse = []
k_list = range(1, 15)
for k in k_list:
    km = KMeans(n_clusters=k)
    km.fit(tsne_df_scale)
    sse.append([k, km.inertia_])
    
tsne_results_scale = pd.DataFrame({'Cluster': range(1,15), 'SSE': sse})
plt.figure(figsize=(12,6))
plt.plot(pd.DataFrame(sse)[0], pd.DataFrame(sse)[1], marker='o')
plt.title('Optimal Number of Clusters using Elbow Method (tSNE_Scaled Data)')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
```

```python
kmeans_tsne_scale = KMeans(n_clusters=4, n_init=100, max_iter=400, init='k-means++', random_state=42).fit(tsne_df_scale)
print('KMeans tSNE Scaled Silhouette Score: {}'.format(silhouette_score(tsne_df_scale, kmeans_tsne_scale.labels_, metric='euclidean')))
labels_tsne_scale = kmeans_tsne_scale.labels_
clusters_tsne_scale = pd.concat([tsne_df_scale, pd.DataFrame({'tsne_clusters':labels_tsne_scale})], axis=1)
tsne_lableddata = pd.concat([finaldata.reset_index(drop=True), pd.DataFrame({'tsne_clusters':labels_tsne_scale})], axis=1)
```

```python
import seaborn as sns
plt.figure(figsize = (15,15))
sns.scatterplot(clusters_tsne_scale.iloc[:,0],clusters_tsne_scale.iloc[:,1],hue=labels_tsne_scale, palette='Set1', s=100, alpha=0.6).set_title('Cluster Vis tSNE Scaled Data', fontsize=15)
plt.legend()
plt.show()
```
